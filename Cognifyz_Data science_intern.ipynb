# Level : 2
#### Tesk : 1,  Table Booking and Online Delivery

- To analyze the table booking and online delivery options among restaurants, we will use the provided dataset to determine the following:

- The percentage of restaurants that offer table booking and online delivery.
- The average ratings of restaurants with and without table booking.
- The availability of online delivery among restaurants with different price ranges.
##### 1. Determine the Percentage of Restaurants Offering Table Booking and Online Delivery. We will calculate the percentage of restaurants that offer table booking and online delivery.

```python
# Total number of restaurants
total_restaurants = len(df)

# Calculate percentage for table booking
table_booking_offered = df['Has Table booking'].value_counts().get('Yes', 0)
percentage_table_booking = (table_booking_offered / total_restaurants) * 100

# Calculate percentage for online delivery
online_delivery_offered = df['Has Online delivery'].value_counts().get('Yes', 0)
percentage_online_delivery = (online_delivery_offered / total_restaurants) * 100

# Print results
print(f"Percentage of restaurants offering table booking: {percentage_table_booking:.2f}%")
print(f"Percentage of restaurants offering online delivery: {percentage_online_delivery:.2f}%")
```

##### 2. Compare the Average Ratings of Restaurants with and without Table Booking.We will compare the average ratings of restaurants that offer table booking with those that do not.

```python
# Calculate average ratings
average_rating_with_booking = df[df['Has Table booking'] == 'Yes']['Aggregate rating'].mean()
average_rating_without_booking = df[df['Has Table booking'] == 'No']['Aggregate rating'].mean()

# Print results
print(f"Average rating with table booking: {average_rating_with_booking:.2f}")
print(f"Average rating without table booking: {average_rating_without_booking:.2f}")
```

##### 3. Analyze the Availability of Online Delivery Among Restaurants with Different Price Ranges.We will analyze how online delivery availability varies among different price ranges.

```python
# Define price range categories
df['Price Range Category'] = pd.cut(df['Average Cost for two'], bins=[0, 500, 1000, 2000, 5000], labels=['Low', 'Medium', 'High', 'Premium'])

# Group by price range and calculate the percentage of online delivery availability
online_delivery_by_price_range = df.groupby('Price Range Category')['Has Online delivery'].value_counts(normalize=True).unstack().fillna(0) * 100

# Print results
print("Availability of online delivery by price range (in %):")
print(online_delivery_by_price_range)
```

#### Summary :

- Percentage of restaurants offering Table Booking 12.12% and Online Delivery 25.66%
- Average rating with Table Booking 3.44 and without Table Booking: 2.56
- Online delivery availability for medium price range food product is better compare to high and low price range food product

# Level :3
#### task : 1 ,Predictive Modeling

- The goal is to build a regression model that predicts the aggregate rating of a restaurant based on various features. We'll follow these steps:



- Model Building: Train different regression models.
- Model Evaluation: Evaluate and compare the models using appropriate metrics.
##### Steps to Achieve the Task:
- Model Building and Training
- Model Evaluation
- Comparison of Models
- We'll use Python and libraries like pandas, scikit-learn, and matplotlib for this task.

```python
# Select features and target variable
features = ['Restaurant Name Length', 'Has Table Booking', 'Has Online Delivery', 'Votes', 'Average Cost for two']
target = 'Aggregate rating'

# Extract features and target
X = df[features]
y = df[target]
```

```python
X
```

```python
y
```

 ##### . Here Create and select Standardize features for the model. Here, we can Split dataset into training and testing sets.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Standardize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)
```

##### . We will build three models: Linear Regression, Decision Tree Regression, and Random Forest Regression.

##### . Linear Regression

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Initialize and train the model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# Predict on the test set
y_pred_lr = lr_model.predict(X_test)
```

#### Decision Tree Regression

```python
from sklearn.tree import DecisionTreeRegressor

# Initialize and train the model
dt_model = DecisionTreeRegressor(random_state=42)
dt_model.fit(X_train, y_train)

# Predict on the test set
y_pred_dt = dt_model.predict(X_test)
```

#### Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor

# Initialize and train the model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predict on the test set
y_pred_rf = rf_model.predict(X_test)
```

```python
sns.displot(y_pred_rf,kind='kde')
```

 ### Model Evaluation
#### Evaluate models using Mean Squared Error (MSE) and R-squared (R²) metrics.

```python
# Define a function to calculate metrics
def evaluate_model(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, r2

# Evaluate Linear Regression
mse_lr, r2_lr = evaluate_model(y_test, y_pred_lr)
print(f'Linear Regression - MSE: {mse_lr}, R²: {r2_lr}')

# Evaluate Decision Tree
mse_dt, r2_dt = evaluate_model(y_test, y_pred_dt)
print(f'Decision Tree - MSE: {mse_dt}, R²: {r2_dt}')

# Evaluate Random Forest
mse_rf, r2_rf = evaluate_model(y_test, y_pred_rf)
print(f'Random Forest - MSE: {mse_rf}, R²: {r2_rf}')
```

##### Summary :
- Data Loading and Preprocessing: Handled missing values and standardized features.
- Feature Engineering: Created binary features and extracted relevant features.
- Model Building: Built Linear Regression, Decision Tree, and Random Forest models.
- Model Evaluation: Evaluated models using MSE and R².
- Comparison: Compared models to identify the best-performing one.

##### Cuisine-Impact Analysis on Ratings,Analyze the impact of different cuisines on restaurant ratings.

```python
# Example dataframe (replace this with your actual dataframe loading step)
df = pd.DataFrame({
    'Cuisines': ['Italian, Chinese', 'Indian', None, 'Mexican, Indian', 123, 'Continental'],
    'aggregate_rating': [4.5, 4.0, 3.5, 4.2, 3.8, 4.1],
    'Votes': [100, 200, 150, 300, 50, 250]
})

# Convert 'Cuisines' to string, replace NaN with an empty string, and drop rows where 'Cuisines' is still NaN
df['Cuisines'] = df['Cuisines'].astype(str).replace('nan', '')
df = df[df['Cuisines'] != '']

# Split cuisines into individual entries for analysis
df['Cuisines'] = df['Cuisines'].str.split(',')

# Explode the dataframe to have one row per cuisine per restaurant
df_exploded = df.explode('Cuisines')

# Strip any leading/trailing spaces from cuisine names
df_exploded['Cuisines'] = df_exploded['Cuisines'].str.strip()

# Calculate the average rating for each cuisine
cuisine_rating = df_exploded.groupby('Cuisines')['aggregate_rating'].mean().reset_index()
cuisine_rating.columns = ['Cuisine', 'Average Rating']

# Display the cuisines with the highest average ratings
print(cuisine_rating.sort_values(by='Average Rating', ascending=False).head(10))
```

```python
# Create a bar plot of the average rating for each cuisine
plt.figure(figsize=(8, 6))
sns.barplot(x='Cuisine', y='Average Rating', data=cuisine_rating.sort_values(by='Average Rating', ascending=False))
plt.xlabel('Cuisine', fontsize=15)
plt.ylabel('Average Rating', fontsize=15)
plt.title('Average Rating for Cuisines', fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

##### Identification of Popular Cuisines,Identify the most popular cuisines based on the number of votes received

```python
# Sum votes for each cuisine
cuisine_votes = df_exploded.groupby('Cuisines')['Votes'].sum().reset_index()
cuisine_votes.columns = ['Cuisine', 'Total Votes']

# Display the cuisines with the most votes
print(cuisine_votes.sort_values(by='Total Votes', ascending=False).head(10))
```

##### Cuisine-Specific Rating Analysis,Determine if there are specific cuisines that tend to receive higher ratings by comparing their average ratings and votes.

```python
# Merge cuisine_rating and cuisine_votes to have a consolidated view
cuisine_analysis = pd.merge(cuisine_rating, cuisine_votes, on='Cuisine')

# Display the consolidated view sorted by average rating
print(cuisine_analysis.sort_values(by='Average Rating', ascending=False).head(10))
```

```python
# Create a bar chart of total votes for each cuisine
plt.figure(figsize=(10, 8))
sns.barplot(x='Cuisine', y='Total Votes', data=cuisine_analysis.sort_values(by='Total Votes', ascending=False))
plt.xlabel('Cuisine', fontsize=15)
plt.ylabel('Total Votes', fontsize=15)
plt.title('Total Votes for Cuisines', fontsize=15)
plt.xticks(rotation=90)
plt.show()
```

#### Conclusion :
- Japanese and French cuisines tend to receive higher ratings, indicating strong customer preference and satisfaction.
- North Indian and Chinese are among the most popular cuisines based on the number of votes, showing high customer engagement and popularity.
- This analysis can be crucial for restaurants to understand trends and make informed decisions regarding menu offerings and marketing strategies.
